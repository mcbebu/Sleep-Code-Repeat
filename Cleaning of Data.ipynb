```python
df1 = pd.read_csv("shipper_address_sample.csv",lineterminator='\n')
df1.head()

df2 = pd.DataFrame()
```

```python
# Concat everything 

df2['new_address'] = df1[df1.columns[2:8]].apply(lambda x: ' '.join(x.dropna().astype(str)),axis=1)
```

```python
df2.head()
```

```python
def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False
    
import math

def float_to_int(float_number):
    return math.floor(float_number)

def unique_words(word):
    new_sentence = word.split(" ")
    new_listing = []
    for element in new_sentence:
        element.strip(" ")
        element=str(element)
        print(type(element))
        if element.isalpha():
            print("This is alpha", element)
            if len(element) <= 1:
                element = element[0].upper()
            else:
                element = element.capitalize()
                
        elif element.isdigit() or element.isnumeric() or element.isdecimal():
            print("This is Element", element)
            element = int(element)
            element = str(element)
            
        elif isfloat(element): 
            element = float(element)
            element = float_to_int(element)
            print("This is float converted", element)
            element = str(element)
        else:
            print("This is not rendered", element)
            
        if element not in new_listing:
            new_listing.append(element)
    
    new_word = " ".join(new_listing)
    return new_word
        

df2['new_address'] = df2['new_address'].apply(unique_words)
```

```python
# Should add some space in front of the # 

def remove_hashtags (text):
    output_str = ""
    for element in text:
        if element != "#":
            output_str += element
    return output_str
            
df2['new_address'] = df2['new_address'].apply(remove_hashtags)
```
