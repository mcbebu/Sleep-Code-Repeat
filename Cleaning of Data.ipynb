```python
# string = re.findall('[A-Z][^A-Z]*',df1['address1'])
# Split it by the Uppercase first 

def split_it_by_uppercase(location):
    # Split by Space first, if it is equals to 1
    x = location.split(" ")
    if len(x) >1: 
        output = re.findall('[A-Z][^A-Z]*',location)
        output = " ".join(output)
        print("This is output:", output)
        
        count = 0
        result_str = ""
        for letters in output:
            if letters.isspace():
                count += 1
            elif not letters.isspace() and count <= 1:
                result_str += letters
                count = 0
            elif not letters.isspace() and count > 1:
                result_str += " " + letters
                count = 0
            
        print("This is result", result_str)
        
        
        return result_str 
    else:
        return location 

df1['address1'] = df1['address1'].apply(split_it_by_uppercase)
```

```python
# Should add some space in front of the # 

def add_a_space_in_front_of_hashtags(word):
    output = ""
    for letter in word:
        if letter == "#":
            output += " #"
        else:
            output += letter
            
    return output 
df1['address1'] = df1['address1'].apply(add_a_space_in_front_of_hashtags)
```
